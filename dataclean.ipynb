```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

DATA CLEANING

```python
data = pd.read_csv("winequality.csv")

data
```

```python
data.shape
```

```python
data = data.drop(["type"],axis=1)
```

```python
data
```

```python
data.isnull().sum()
```

```python
numeric_columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides','pH', 'sulphates']
```

```python
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].mean())
```

EXPLORATORY DATA ANALYSIS

```python
data = pd.read_csv('wine quality.csv')
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='quality', palette='viridis')
plt.title('Wine Quality Distribution')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = data.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.pairplot(data[['fixed acidity', 'volatile acidity', 'citric acid', 'alcohol', 'quality']], hue='quality', palette='viridis')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='quality', y='alcohol', palette='viridis')
plt.title('Alcohol Content by Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Alcohol Content')
plt.show()
```
